```python
import json

# Define the columns to include
columns = [
    'flatshare_id', 'available', 'bills_included', 'couples_ok', 'deposit', 
    'detail', 'furnishings', 'gender', 'living_room', 'location', 
    'minimum_term', 'occupation', 'num_flatmates', 'parking', 'pets_ok', 
    'postcode', 'property_type', 'rent', 'room_type'
]

# Create new dataframe with selected columns
new_df = df[columns].copy()

# Add a json_data column containing all column values as a JSON string
new_df['json_data'] = new_df.apply(lambda row: json.dumps(row.to_dict()), axis=1)
new_df['images'] = df['images'].apply(lambda x: json.loads(x))
new_df.head()
```

```python
# Print the full JSON string for flatshare_id 18069392
json_text = new_df.loc[new_df['flatshare_id'] == 18069392, 'json_data'].values
if len(json_text) > 0:
    print(json_text[0])
else:
    print("No entry found for flatshare_id 18069392")
```

```python
new_df.to_csv('listings_clean.csv', index=False)
```

```python
from redis.commands.search.field import VectorField, TextField
from redis.commands.search.indexDefinition import IndexDefinition

DIMENSION = 1536

schema = (
    TextField("flatshare_id"),
    TextField("bills_included"),
    TextField("furnishings"),
    TextField("images"),
    TextField("postcode"),
    TextField("rent"),
    TextField("room_type"),
    TextField("property_type"),
    TextField("detail"),
    VectorField(
        "detail_embedding",
        "HNSW",
        {
            "TYPE": "FLOAT32",
            "DIM": DIMENSION,
            "DISTANCE_METRIC": "COSINE"
        }
    )
)

# Create index on all keys that start with "doc:"
r.ft("idx_docs").create_index(
    schema,
    definition=IndexDefinition(prefix=["doc:"])
)

    # 'flatshare_id', 'available', 'bills_included', 'couples_ok', 'deposit', 
    # 'detail', 'furnishings', 'gender', 'living_room', 'location', 
    # 'minimum_term', 'occupation', 'num_flatmates', 'parking', 'pets_ok', 
    # 'postcode', 'property_type', 'rent', 'room_type'
```
